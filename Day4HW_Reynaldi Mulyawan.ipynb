```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create Dataframe
df = pd.DataFrame(([["Anna", 45.2, 155, 2000, "female", "New York"],
                     ["Dane", 50.5, 160, 1997, "male", "Chicago"],
                     ["David", 62.1, 162, 1996, "male", "Austin"],
                     ["Kevin", 64.0, 170, 1987, "male", "New York"],
                     ["Joe", 59, 167, 1998, "male","NaN"],
                     ["Rose", np.nan, 159, 2003, "female", "New York"]]),
                     columns = ["name", "weight", "height", "birth_year", "gender", "city"])
df
```

```python
#Create another Dataframe
df2 = pd.DataFrame(([["Anna", "Anna Levis"],
                     ["David", "David John"],
                     ["Dane", "Dane Sebastian"],
                     ["Rose", "Rose Kimberly"]]),
                     columns = ["nick_name", "full_name"])
df2
```

```python
#Merge Two Dataframe using Merge
df3 = df.merge(df2, how = "outer", left_on = "name", right_on = "nick_name").drop(columns = "nick_name")
```

```python
df3
```

```python
#Creating BMI using Column Operation
df3['BMI'] = (df3['weight']/((df3['height']/100)**2))
df3
```

```python
#Creating Conditional for BMI Classification
conditions = [
    (df3['BMI'] < 18.5),
    (df3['BMI'].between(18.5, 25)),
    (df3['BMI'].between(25, 30)),
    (df3['BMI'].between(30, 35)),
    (df3['BMI'].between(35, 40)),
    (df3['BMI'] > 40)]
choices = ['underweight', 'normal weight', 'overweight', 'obesity class 1','obesity class 2','obesity class 3']
df3['BMI_class'] = np.select(conditions, choices)
df3
```
